<h1>Microsoft Corporation (MSFT)</h1>

```python
import matplotlib.pyplot as plt
import yfinance as yf
```

<h4>MSFT Stock Price</h4>

```python
data = yf.Ticker('MSFT')
price = data.history(period='10y')
```

```python
price['Close'].plot()
plt.show()
```

<h4>MSFT Daily Return</h4>

```python
pct_change = price['Close'].pct_change()
pct_change.plot()
plt.show()
```

```python
pct_change.plot(kind="hist")
plt.show()
```

<h4>1$ Investment Return</h4>

```python
returns = (pct_change + 1).cumprod()
returns.plot()
plt.show()
```
