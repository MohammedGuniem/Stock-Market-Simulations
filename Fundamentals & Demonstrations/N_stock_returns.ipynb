<h1>Stock Return</h1>

```python
from IPython.display import display
import matplotlib.pyplot as plt
import yfinance as yf
```

<h5>We will start by calculating the daily returns of a stock.</h5>

<p>Let's get the stock prices for Tesla over the past year:</p>

```python
data = yf.Ticker('TSLA')
price = data.history(period='10y')
display(price)
```

<h5>In order to calculate the daily returns, we will use the pct_change() function</h5>
<p>Which calculates the percentage change between the current element and a prior one.</p>

<p>We will use it on the 'Close' column:</p>

```python
pct_change = price['Close'].pct_change()
display(pct_change)
```

<h5>To visualize the results, we can create a plot for the daily returns:</h5>

```python
pct_change.plot()
plt.show()
```

<h5>We can also make a histogram to see the distribution:</h5>

```python
pct_change.plot(kind="hist")
plt.show()
```

<h5>After understanding how the returns are distributed, we can calculate the returns from an investment</h5>
<p>For that, we need to calculate the cumulative returns, which can be done using the cumprod() function</p>

```python
returns = (pct_change + 1).cumprod()
display(returns)
```

<h5>Here is the plot:</h5>
<p>The plot shows how a $1 investment would grow.</p>

```python
returns.plot()
plt.show()
```
