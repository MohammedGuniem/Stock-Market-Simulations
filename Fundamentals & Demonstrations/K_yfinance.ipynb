<h1>yfinance</h1>

```python
from IPython.display import display
import matplotlib.pyplot as plt
import yfinance as yf
import pprint
```

<h4>The Ticker module allows us to access company data based on their market ticker symbol:</h4>

```python
data = yf.Ticker("TSLA")
```

<h4>Now, we can access the company information under the corresponding fields:</h4>

```python
pp = pprint.PrettyPrinter(indent=2)
pp.pprint(data.info)
```

<h4>Let's output the profit margins and RoE:</h4>

```python
print("profitMargins = ", data.info['profitMargins'])
print("returnOnEquity = ", data.info['returnOnEquity'])
```

<h4>In addition to the info fields, the data object provides the following fields:</h4>

<h5>dividends</h5>

```python
display(data.dividends)
```

<h4>splits</h4>

```python
display(data.splits)
```

<h4>balance_sheet</h4>

```python
display(data.balance_sheet)
```

<h4>cashflow</h4>

```python
display(data.cashflow)
```

<h4>earnings_dates</h4>

```python
display(data.earnings_dates)
```

<h4>plot the cash flow table as bar charts</h4>

```python
x = data.earnings_dates
x.plot(kind="bar")
plt.show()
```
