<h1>Multiple Stocks</h1>

```python
from IPython.display import display
import matplotlib.pyplot as plt
import yfinance as yf
```

<h4>Let's take 3 different tickers and use the pct_change() function to calculate the daily returns:</h4>
<p>The code below gets the stock prices for the given stocks and applies to pct_change() function to the Close price.</p>

```python
data = yf.download("AAPL MSFT TSLA", start='2021-01-01') 
x = data['Close'].pct_change()
display(x) 
```

<h4>To better understand the return values, we can use the describe() function on the DataFrame to get the descriptive statistics:</h4>

```python
display(x.describe()) 
```

<p>Descriptive statistics include the mean, standard deviation, min, max values, as well as the 25/50/75th% percentiles.</p>

<h4>Let's visualize the results with charts.</h4>
<p></p>Here is a chart comparing the daily prices of the stocks:</p>

```python
data['Close'].plot()
plt.show() 
```

<h4>We can also plot their daily returns:</h4>

```python
x.plot()
plt.show() 
```

<h4>Here are the cumulative returns for the given stocks:</h4>

```python
(x +1).cumprod().plot()
plt.show()
```

<p>Visualization helps to understand how the stocks perform in the given period.</p>
