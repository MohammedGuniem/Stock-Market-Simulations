<h1>Web Scraping</h1>

<p>
    For example, let's scrape the list of S&P 500 companies from Wikipedia.<br/>
    The list is available on the Wikipedia article page as a table.<br/>
    We simply need to call the read_html() function with the URL of the page as the parameter:
</p>

<p>
    Each table on the web page is stored as the DataFrame at a separate index.<br/>
    The first table has the index 0, the second table - the index 1, and so on.<br/>
    Let's access and output the first table:
</p>

```python
import pandas as pd
from IPython.display import display
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
df = data[0]
display(df)
```

<p>Let's select and output only the Symbol and Security columns:</p>

```python
display(df[df['Security'] == 'Apple Inc.'])
```

<p>The info() function can be used to see all available columns:</p>

```python
print(df.info())
```
