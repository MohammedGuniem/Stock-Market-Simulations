<h1>Stock Correlations</h1>

<h4>In finance, correlation is a statistic that measures the degree to which two securities move in relation to each other.</h4>
<p>We can easily calculate the correlations between stocks in Python, with the corr() function.</p>

```python
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
data = yf.download("AMZN AAPL NFLX GOOG NVDA", start='2020-01-01') 
x = data['Close'].pct_change()
corr = x.corr()
display(corr) 
```

<h4>Let's draw a nice graphic using the statmodels package:</h4>
<p>This will result in a graphic that shows the correlations between the given stocks in the given period.</p>

```python
sns.heatmap(corr, annot=True, cmap='PiYG', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

<h4>The corr() function results in a matrix that includes values for each stock pair.</h4>
<p>The values are in the range of -1 to 1.</p>
<p>A positive correlation means that the stocks have returns that are positively correlated and move in the same direction.</p>
<p>+1 means that the returns are perfectly correlated.</p>
<p>A correlation of 0 shows no relationship between the pair.</p>
<p>A negative correlation shows that the returns move in different directions.</p>
<h4>Finding stocks that have low correlation helps to diversify an investment portfolio and minimize risk.</h4>
