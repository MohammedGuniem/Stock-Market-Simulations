<h1>Access Stock Prices</h1>

```python
from IPython.display import display
import matplotlib.pyplot as plt
import yfinance as yf
```

<h5>yfinance also provides the stock prices of the given ticker.</h5>

<p>This will output the stock prices for the last month.</p>

```python
data = yf.Ticker('TSLA')
display(data.history())
```

<p>We can provide a period parameter:</p>

```python
display(data.history(period='5d'))
```

<p>Other valid periods are: 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max.</p>

<p>We can also provide custom start and end dates:</p>

```python
display(data.history(start='2021-01-01', end='2021-06-30'))
```

<p>Let's plot the daily Close price of Tesla stock for the last month:</p>

```python
data = yf.Ticker('TSLA')
x = data.history()['Close']
x.plot()
plt.show()
```

<p>Output the average Close price of Tesla stock for the last 3 months.</p>

```python
data = yf.Ticker('TSLA')
x = data.history('3mo')['Close']
print(x.mean())
```

<h5>In addition, yfinance allows you to download historical prices for more than one stock simultaneously.</h5>

<p>For example, let's take the stock prices of Apple, Microsoft and Tesla with one line:</p>

```python
data = yf.download("AAPL MSFT TSLA", start='2021-01-01')
display(data['Close'])
```

<p>Now we can plot the stock prices of all the 3 tickers:</p>

```python
data['Close'].plot()
plt.show()
```
