<h1>Plotting data</h1>

<p>First we import the matplotlib package</p>

```python
import matplotlib.pyplot as plt
```

<p>Let's assume we have a company's 5 month revenue data in an array:<br/>
We want to plot a chart visualizing the revenue data.<br/>

<p>To do that, we simply need to call the plot function on our data:</br>
You can call the savefig function to save the previous plot before showing the plot in this notebook.</br>
Then you can show the plot on this notebook using the show function</p>

```python
rev = [18000, 25000, 20000, 45000, 32000]
plt.plot(rev)
plt.savefig('plots/plot_saved.png')
plt.show()
```

<p>The plot() function can also take values for both the x and y axis.<br/>
Let's add the month names on the horizontal axis:</p>

```python
rev = [18000, 25000, 20000, 45000, 32000]
months = ['June', 'July', 'August', 'September', 'October']
plt.plot(months, rev)
plt.show()
```
