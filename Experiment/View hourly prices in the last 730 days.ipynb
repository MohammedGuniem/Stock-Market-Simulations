```python
import matplotlib.pyplot as plt
from datetime import datetime
import yfinance as yf
import seaborn as sns
import pandas as pd
import numpy as np
import os
```

```python
# Download historical stock data for a specific ticker
symbol = 'AAPL'

stock = yf.Ticker(symbol)

stock_data = stock.history(period="730d", interval="1h")
```

```python
display(stock_data)
```

```python
print("Number of open trading days in market: ", len(stock_data))
```

```python
import matplotlib.pyplot as plt

# Creating the plot
plt.plot(stock_data.index, stock_data["Close"].values)

# Adding title and labels
plt.title("Close vs. Datetime")
plt.xlabel("Datetime")
plt.ylabel("Close")

# Rotate x-tick labels vertically
plt.xticks(rotation=90)

# Enabling grid
plt.grid(True)

# Display the plot
plt.show()
```
